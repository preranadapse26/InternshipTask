```python
from sklearn.datasets import load_breast_cancer
import pandas as pd

# Load dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)

# Display the first few rows
print("Dataset shape:", df.shape)
df.head()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Apply PCA to reduce dimensions to 2
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df)

# Create a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components, columns=['Principal Component 1', 'Principal Component 2'])

# Add target labels
pca_df['Target'] = data.target

# Plot the PCA result
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['Principal Component 1'], pca_df['Principal Component 2'], c=pca_df['Target'], cmap='coolwarm')
plt.title('PCA: Breast Cancer Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Target (0 = Malignant, 1 = Benign)')
plt.show()
```

```python
from sklearn.manifold import TSNE

# Apply t-SNE to reduce dimensions to 2
tsne = TSNE(n_components=2, perplexity=30, random_state=42)
tsne_components = tsne.fit_transform(df)

# Create a DataFrame with the t-SNE components
tsne_df = pd.DataFrame(data=tsne_components, columns=['t-SNE Component 1', 't-SNE Component 2'])
tsne_df['Target'] = data.target

# Plot the t-SNE result
plt.figure(figsize=(8, 6))
plt.scatter(tsne_df['t-SNE Component 1'], tsne_df['t-SNE Component 2'], c=tsne_df['Target'], cmap='coolwarm')
plt.title('t-SNE: Breast Cancer Dataset')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.colorbar(label='Target (0 = Malignant, 1 = Benign)')
plt.show()
```
