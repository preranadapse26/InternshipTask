Using Pyplot

```python
import plotly.express as px
import pandas as pd

```

```python
df = px.data.iris()
```

```python
fig = px.scatter(
    df,
    x='sepal_width',
    y='sepal_length',
    color='species',
    size='petal_length',
    hover_data=['petal_width']
)
```

```python
fig.update_layout(
    title='Interactive Iris Dataset Scatter Plot',
    xaxis_title='Sepal Width',
    yaxis_title='Sepal Length'
)
```

```python
fig.show()
```

Using Seaborn


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('tips')
```

```python
# Create a pairplot with regression lines
sns.pairplot(df, hue='sex', kind='reg', diag_kind='kde', palette='coolwarm')
plt.show()
```
